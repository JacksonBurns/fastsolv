```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
```

```python
aqsoldb = pd.read_csv(r"https://raw.githubusercontent.com/mcsorkun/AqSolDB/master/results/data_curated.csv", usecols=["Solubility"])
bigsoldb = pd.read_csv(Path("../data/krasnov/bigsoldb_chemprop_nonaq.csv"), usecols=["logS"])
```

```python
df = pd.concat((aqsoldb, bigsoldb), axis=1)
df.columns = ["AqSolDB", "BigSolDB-nonaq"]
```

```python
# Plot the histograms
plt.hist(df['AqSolDB'], bins=25, color='blue', alpha=0.2, density=True, label='AqSolDB', edgecolor='black', rwidth=0.9)
plt.hist(df['BigSolDB-nonaq'], bins=25, color='green', alpha=0.2, density=True, label='BigSolDB-nonaq', edgecolor='black', rwidth=0.9)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('AqSolDB vs. BigSolDB-nonaq logS Distribution')

# Show legend
plt.legend()

# Show plot
plt.show()
```
